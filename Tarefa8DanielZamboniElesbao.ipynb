**1)** O índice de massa corporal (IMC) de uma pessoa é calculado por:

$$IMC = \frac{m}{h^2} $$

em que $m$ é a massa corporal, em kg, e $h$ é a altura, em metros.

O IMC tem a seguinte classificação:

- menor que 17 kg/m$^2$ -- Muito abaixo do peso
- 17 a 18,5 kg/m$^2$ -- Abaixo do peso
- 18,5 a 25 kg/m$^2$ -- Peso normal
- 25 a 30 kg/m$^2$ -- Acima do peso
- 30 a 35 kg/m$^2$ -- Obesidade Grau I
- 35 a 40 kg/m$^2$ -- Obesidade Grau II
- maior que 40 kg/m$^2$ -- Obesidade Grau III

Escreva um código em Python para calcular o IMC e mostrar a classificação do IMC. Teste para a) $m = 52$ kg e $h = 1,58$ m, b) $m = 83$ kg e $h = 1,75$ m, c) seu peso e altura.

```python
#Calcula o IMC e mostra a classificação do IMC de acordo com o exercício.
import numpy as np #importa a biblioteca numpy como np
m = [ 52, 83, 95 ] #define a array m contendo as massas de acordo com os dados do exercício
h = [1.58, 1.75, 1.78] #define a array h contendo as massas de acordo com os dados do exercício
for massa, altura in zip(m,h): #cria um laço que percorre as arrays m e h em paralelo e define as variáveis massa e altura com o conteúdo de cada índice das arrays
  imc = massa/altura**2 #calcula o IMC
  if imc < 17: #testa o IMC para imprimir a sua classificação
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Muito abaixo do peso') #retorna os dados
  elif imc>=17 and imc < 18.5:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Abaixo do peso')
  elif imc>=18.5 and imc < 25:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Peso normal')
  elif imc>=25 and imc < 30:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Acima do peso')
  elif imc>=30 and imc < 35:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Obesidade Grau I')
  elif imc>=35 and imc < 40:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é {imc:.2f} e sua classificação é Obesidade Grau II')
  else:
    print (f'O IMC de uma pessoa de massa {massa:.2f} kg, e altura {altura:.2f} metros é: {imc:.2f} e sua classificação é Obesidade Grau III')
```

**2)** A acidez $A(x)$ de uma solução de hidróxido de magnésio em ácido clorídrico, sob certas condições experimentais, é dada pela equação:

$$A(x) = x^3 + 3x^2 − 54$$
na qual $x$ é a concentração de íons hidrônio.

Pede-se:

1) Gere o gráfico de $A(x)$ em função de $x$ para 0 $\leqslant x \leqslant 8$.

2) Determine a concentração $x$ do íon de hidrônio que resulta em solução saturada (i.e., com acidez nula). Acrescente uma instrução que gere um ponto vermelho no gráfico correspondente à acidez nula da solução.

```python
#Gera o gráfico da acidez de uma solução de hidróxido de magnésio em ácido clorídrico de acordo com os dados do exercício
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
x = np.arange(0, 8, 0.1) #define x como o intervalo entre 0 e 8
A = x**3 + 3*x**2 - 54 #fórmula da acidez fornecida pelo exercício
plt.figure() #executa função que desenha o gráfico
plt.title('Gráfico da acidez de uma solução')#Título do Gráfico
plt.plot(x, A) #cria o gráfico de A em função de x.
plt.plot(x[A==0], 0, marker='.', color='red') #coloca o ponto na intersecção onde a acidez é 0.
plt.annotate('Acidez nula',(x[A==0], 0), xytext=(2,200), 
  arrowprops=dict(facecolor='red')) # insere uma anotação no gráfico indicando o ponto onde a acidez é nula.
plt.xlabel('íons hidrônio') #define o significado do eixo x como íons hidrônio
plt.ylabel('Acidez') #define o significado do eixo y como Acidez
plt.show() #exibe o gráfico
```

**3)**  Faça o gráfico de:
$$f(x)=|x−2|+|2x+1|−x−6$$

Faça o gráfico com $x$ variando de -10 a 10.

Mostre o menor valor de $x$ tal que $f(x)>0$ e $x>0$.

```python
#Gera o gráfico de uma função de acordo com o exercício
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
x = np.arange(-10, 10, 0.001) #define x como o intervalo entre -10 e 10
f = abs(x-2) + abs(2*x+1) - x - 6 #fórmula fornecida pelo exercício
menor=np.min(x[(f>0) & (x>0)]) #Calcula menor valor de x para f(x) > 0 e x > 0 e o atribui à variável 'menor'
print (f'O menor valor de x para f(x) > 0 e x > 0 é {menor:.2f}') #imprime a variável menor
plt.figure() #executa função que desenha o gráfico
plt.plot(x, f) #cria o gráfico de f em função de x.
plt.plot(menor, 0.01, marker='.', color='red') #coloca o ponto no menor valor de x para f(x) > 0 e x > 0.
plt.annotate('Menor x para f(x) > 0',(menor, 0.001), xytext=(0,10), 
  arrowprops=dict(facecolor='red')) # insere uma anotação no gráfico indicando o significado do ponto.
plt.show() #exibe o gráfico
```

**4)** Faça o gráfico da seguinte função:

$$f(x) = x^2 – \mathrm{sen}(0.784x^2) – 2$$

Mostre as raízes da função (i.e. os valores de $x$ para os quais $f(x) = 0$).

```python
#Gera o gráfico de uma função de acordo com o exercício
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
x = np.arange(-10, 10, 0.001) #define x como o intervalo entre -10 e 10
f = x**2 - np.sin(0.784*(x**2)) - 2 #fórmula fornecida pelo exercício
raizes = x[np.isclose(f,0, atol=0.002)] #Calcula as raízes da função que não tem valor exato
print (f'As raízes da função são aproximadamente {raizes[0]:.3f} e {raizes[1]:.3f}') #imprime o valor aproximado das raízes da função
plt.figure() #executa função que desenha o gráfico
plt.plot(x, f) #cria o gráfico de f em função de x.
plt.plot(raizes[0],0, marker='.', color='red') #coloca o ponto na primeira raiz.
plt.plot(raizes[1],0, marker='.', color='red') #coloca o ponto na segunda raiz.
plt.annotate('Primeira Raiz',(raizes[0],0), xytext=(-5,40),
  arrowprops=dict(facecolor='red')) # insere uma anotação no gráfico indicando o significado do ponto.
plt.annotate('Segunda raiz', (raizes[1],0), xytext=(0,20),
  arrowprops=dict(facecolor='red')) # insere uma anotação no gráfico indicando o significado do ponto.
plt.show() #exibe o gráfico
```
