```python
#Cria uma função que calcula um vetor e exibe o gráfico
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
def vetorar ( t , A , b, c ): #define a função que recebe os valores,  t , A , b e c
  x = A * np.sqrt( 1 + b * t ** 2 ) + c #cálculo conforme a fórmula
  plt.plot ( t , x, label = f'A={A}, b={b} e c={c}' ) #cria o gráfico de x em função de t.
  return x #retorna a variável x
A = [ 2 , 10 , -3 ] #cria arrays com os valores a serem testados de acordo com o enunciado
b = [ 0.5 , 0.2 , -1.5 ]
c = [ 0 , 1 , -10 ]
t = [np.arange( 0, 10, 0.001 ) , np.arange( 0, 15, 0.2 ) , np.arange( -0.5, 0.5, 0.001 ) ]
plt.figure( ) #executa função que desenha o gráfico
for item in range (0,3): #itera pelos valores das arrays
  vetorar ( t[item] , A[item], b[item] , c[item] ) #chama a função informando como valores de entrada os números das arrays criadas
plt.title( 'Vetoração' ) #título do gráfico
plt.legend ( ) #exibe as legendas
plt.show( ) #exibe o gráfico
```

```python
#Cria uma função para calcular a renda média de um time no Campeonato Brasileiro 2018.
import pandas as pd #importa a biblioteca pandas como pd
brasileirao = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')#Lê o CSV fornecido pelo exercício
def renda_media ( time, jogos ): #define a função que recebe o nome do time e o Dataframe
  media = brasileirao.query("Mandante == @time")['Renda (R$)'].mean() #calcula a renda média dos jogos do time como mandante
  print (f'\nA Renda Média em jogos do {time} como mandante foi R$ {media:.2f}') #exibe a média calculada formatada
times = ['Santos', 'São Paulo', 'Corinthians', 'Palmeiras'] #cria uma array com os times a serem testados
for time in times: #chama a função para cada item da array
  renda_media (time, brasileirao) 
```

```python
#Cria uma função que recebe o valor de dois pontos, encontra a inclinação da reta e o ponto em que a reta cruza o eixo y
#e desenha o gráfico da reta encontrada com N pontos, além de mostrar os dois pontos dados como entrada, indicados com marcadores quadrados.
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
def retas ( x1, y1, x2, y2, N): #define a função que recebe os valores das coordenadas dos pontos e da quantidade de pontos da reta a ser desenhada
  m = (y2 - y1) / (x2 - x1) #calcula a inclinação da reta
  b = y1 - m * x1 #calcula onde a reta cruza o eixo y
  plt.figure( ) #executa função que desenha o gráfico
  plt.plot ([y1,y2] , [x1,x2], linestyle = (0,(N,1)), label = f'Inclinação "m": {m:.2f}') #desenha a reta com N pontos e legenda contendo o valor da inclinação m
  plt.plot( b,0, marker='o', label = f'Ponto "b" [y=0]: {b:.2f}') #marca o ponto b onde y é 0
  plt.plot( y1, x1, marker='s', label = f'[{x1}, {y1}]') #marca o ponto 1
  plt.plot( y2, x2, marker='s', label = f'[{x2}, {y2}]') #marca o ponto 2
  plt.legend () #cria as legendas
  plt.show( ) #exibe o gráfico
  return m, b
x1 = [ -19 , -2 , 7 ] #cria arrays com os valores a serem testados de acordo com o enunciado
y1 = [ 2 , 8 , 13 ]
x2 = [ 10 , 20 , 1 ]
y2 = [ -10 , 43 , 2 ]
N = [ 1000 , 2000 , 500 ]
for item in range (0,3): #itera pelos valores das arrays
  retas ( x1[item] , y1[item], x2[item] , y2[item], N[item] ) #chama a função informando como valores de entrada os números das arrays criadas
```
