**1)** A tensão elétrica medida em um circuito, em volts, tem a seguinte expressão no instante $t$, em segundos:

$$v(t) = e^{-0.5t}\cos\left(2\pi0.8t\right)$$

- Faça o gráfico da tensão em função do tempo.

Sugestão: fazer o gráfico com tempo entre 0 e 10 s com incremento de 0.05 s.

```python
#Cria o gráfico da tensão elétrica medida em um circuito, em volts conforme dados fornecidos pelo exercício 
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
t = np.linspace(0, 10, 200) #define t como o intervalo entre 0 e 10 a cada 0.05 segundos
v = np.exp(-0.5*t)*np.cos(2*np.pi*0.8*t) #fórmula da tensão elétrica do circuito conforme dados do exercício
plt.figure()
plt.plot(t, v) #cria o gráfico de v em função de t
plt.xlabel('tempo (s)') #define o significado do eixo x como tempo em segundos
plt.ylabel('tensão') #define o significado do eixo y como tensão
plt.show() #exibe o gráfico
```

**2)** Faça um gráfico da equivalência da temperatura em Fahrenheits em função da temperatura em Celsius:

$$F = \frac{9}{5}C + 32 $$

Faça o gráfico tomando 100 pontos entre -20 $^\circ$C e 100 $^\circ$C.

```python
#Desenha o gráfico da equivalência entre as medidas de temperaturas Celsius e Fahrenheit 
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
C = np.linspace(-20, 100, 100) #define C como a temperatura em celsius no intervalo entre -20° C e 100° C com 100 subdivisões do intervalo
F = (9/5*C)+32 #fórmula da conversão Celsius/Fahrenheit
plt.figure()
plt.plot(C,F) #cria o gráfico de F em função de C.
plt.xlabel('temperatura (C)') #define o significado do eixo x como temperatura em graus Celsius
plt.ylabel('temperatura (F)') #define o significado do eixo y como temperatura em graus Fahrenheit
plt.show() #exibe o gráfico
```

**3)** Entre os anos de 1994 e 2022, a população de uma cidade cresce aproximadamente de acordo com a seguinte função:

$$P(a) = 800000e^{\frac{a - 1994}{40}}$$

em que $a$ é o ano e $P$ é o número de habitantes.

- Faça o gráfico do número de habitantes entre 1994 e 2022.

```python
#Cria o gráfico do desenvolvimento populacional de uma cidade de acordo com a formula fornecida no exercício.
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
a = np.linspace(1994, 2022, 29) #define a como o intervalo de anos entre 1994 e 2022
P = 800000*np.exp((a-1994)/40) #fórmula do crescimento populacional dada pelo exercício
plt.figure()
plt.plot(a, P) #cria o gráfico de P em função de a
plt.xlabel('ano') #define o significado do eixo x como ano
plt.ylabel('população') #define o significado do eixo y como população
plt.show() #exibe o gráfico
```
