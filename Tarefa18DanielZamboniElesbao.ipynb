```python
#Gráfico de dispersão do tamanho do pé (FootLen) em função da altura da pessoa (Height). Calcula a correlação e plota a reta de regressão entre esses dados.
import pandas as pd #importa a biblioteca pandas como pd
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
info_postura = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/BDSinfo.csv', sep='\t')#Lê o CSV fornecido pelo exercício
mediaA = np.mean(info_postura['Height'])
mediaP = np.mean(info_postura['FootLen'])
desvioA = info_postura['Height'] - mediaA
desvioP = info_postura['FootLen'] - mediaP
m = np.sum(desvioA*desvioP)/np.sum(desvioA**2) #define as variáveis e calcula o valor da regressão
b = mediaP - m*mediaA # não foi possível utilizar a função np.polyfit pois há dados inválidos no DataFrame
corr_pe_altura = info_postura['FootLen'].corr( info_postura['Height'] ) #Calcula a correlação
plt.figure( ) #executa função que desenha o gráfico
plt.title('Correlação entre tamanho do pé (FootLen) em função da altura da pessoa (Height)') #Título do Gráfico
plt.plot ( info_postura['Height'], info_postura['FootLen'] , marker='o', color='red', linestyle='', label=f'Correlação: {corr_pe_altura:.2f}')#Cria o Gráfico de dispersão
plt.plot(info_postura['Height'], m*info_postura['Height']+b, linestyle = '-', color = 'blue')    
plt.xlabel('Altura') #define o eixo x como Altura
plt.ylabel('Tamanho do Pé') #define o eixo y como Tamanho do pé
plt.legend ( ) #gera as legendas
plt.grid() #adiciona grade ao gráfico para melhor visualização
plt.show( ) #exibe o gráfico
```

```python
#Gráfico de dispersão entre o crescimento do PIB e a inflação anual de 1961 a 2018.
import pandas as pd #importa a biblioteca pandas como pd
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
inflaAnual = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflaAnual.csv')#Lê o CSV fornecido pelo exercício
pibAnual =  pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/pibAnual.csv')#Lê o CSV fornecido pelo exercício
corr_inf_pib = pibAnual.query('Ano>=1968')['Variação anual do PIB real (%)'].corr(inflaAnual.query('Ano>=1968')['Inflação']) #calcula a correlação
i, p = np.polyfit(inflaAnual.query('Ano>=1968')['Inflação'], pibAnual.query('Ano>=1968')['Variação anual do PIB real (%)'], deg=1) #Calcula a reta de regressão
plt.figure( ) #executa função que desenha o gráfico
plt.title('Correlação entre taxa de crescimento do PIB e Inflação') #Título do Gráfico
plt.plot (pibAnual.query('Ano>=1968')['Variação anual do PIB real (%)'], inflaAnual.query('Ano>=1968')['Inflação'],
          marker='o', color='red', linestyle='', label=f'Correlação: {corr_inf_pib:.2f}')#Cria o Gráfico da Renda pelo Público
plt.plot(pibAnual.query('Ano>=1968')['Variação anual do PIB real (%)'], 
         i * pibAnual.query('Ano>=1968')['Variação anual do PIB real (%)'] + p) #cria a reta de regressão
plt.xlabel('Taxa de Crescimento do PIB') #define o eixo x como Taxa de Crescimento do PIB
plt.ylabel('Inflação') #define o eixo y como Inflação
plt.legend ( ) #gera as legendas
plt.grid() #adiciona grade ao gráfico para melhor visualização
plt.show( ) #exibe o gráfico
```

```python
#Gráfico de dispersão do número de medicamentos que a pessoa toma (Nmedication) em função da idade da pessoa (Age). 
#Calcula a correlação e plota a reta de regressão entre esses dados.
import pandas as pd #importa a biblioteca pandas como pd
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
info_postura = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/BDSinfo.csv', sep='\t')#Lê o CSV fornecido pelo exercício
corr_med_idade = info_postura['Nmedication'].corr( info_postura['Age'] ) #Calcula a correlação
i, m = np.polyfit( info_postura['Age'],info_postura['Nmedication'], deg=1) #Calcula a reta de regressão
plt.figure( ) #executa função que desenha o gráfico
plt.title('Correlação entre número de medicamentos que a pessoa toma em função da idade') #Título do Gráfico
plt.plot ( info_postura['Age'], info_postura['Nmedication'], marker='o', color='red', linestyle='', label=f'Correlação: {corr_med_idade:.2f}')#Cria o Gráfico de dispersão
plt.plot( info_postura['Age'], i*info_postura['Age']+m, linestyle = '-', color = 'blue') #desenha a reta de regressão
plt.xlabel('Idade') #define o eixo x como Idade
plt.ylabel('Núemro de medicamentos') #define o eixo y como Número de medicamentos
plt.legend ( ) #gera as legendas
plt.grid() #adiciona grade ao gráfico para melhor visualização
plt.show( ) #exibe o gráfico
```
