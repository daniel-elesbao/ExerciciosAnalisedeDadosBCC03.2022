```python
#Estatísticas de força e torque medidos no chão quando uma pessoa fica em pé por um minuto.
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
import numpy as np #importa a biblioteca numpy como np
import pandas as pd #importa a biblioteca pandas como pd
balance = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/balance.csv')#Lê o CSV fornecido pelo exercício
#display (balance)
plt.figure() #executa função que desenha o gráfico
plt.title('Torque na direção "y" pelo tempo')#Título do Gráfico
plt.plot (balance['Time[s]'],balance['My[Nm]']) #desenha o gráfico de My sobre Time.
plt.xlabel ('Tempo') #Cria uma legenda para o eixo x
plt.xlabel ('Torque em y') #Cria uma legenda para o eixo x
plt.show() #exibe o gráfico
media = balance['My[Nm]'].mean()#Define a variável media com a média dos valores de torque no eixo y
desvio = balance['My[Nm]'].std()#Define a variável desvio com o desvio-padrão dos valores de torque no eixo y
print (f'\n\nA média do torque no eixo y é {media} Nm e o desvio padrão é {desvio}') #Exibe média e desvio-padrão formatados
```

```python
#Estatísticas da inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019 (dados disponibilizados por Ipeadata).
import pandas as pd #importa a biblioteca pandas como pd
igpdi = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv') #importa os dados do csv informado no exercício
marco = igpdi.query("Mês == 3")#Cria um data frame com os dados para o mês de março
mediana = marco['Inflação'].median()#Define a variável mediana contendo a mediana da inflação do dataframe criado.
media = marco['Inflação'].mean()#Define a variável media contendo a média da inflação do dataframe criado.
desvio = marco['Inflação'].std()#Define a variável desvio contendo o desvio-padrão da inflação do dataframe criado.
histo =  marco['Inflação'].hist()#Define a variável histo como o resultado do comando que cria o histograma dos dados de inflação do dataframe criado.
print (f'A mediana da inflação mensal nos meses de março foi {mediana}\nA média da inflação mensal nos meses de março foi {media}')
print (f'O desvio-padrão da inflação mensal nos meses de março foi {desvio}\n\nHistograma dos meses de março:')
display (histo) #exibe os resultados
```

```python
#Estatísticas da inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019 (dados disponibilizados por Ipeadata).
import pandas as pd #importa a biblioteca pandas como pd
igpdi = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv') #importa os dados do csv informado no exercício
marco = igpdi.query("Mês == 3 and Ano>=1995")#Cria um data frame com os dados para o mês de março a partir de 1995
mediana = marco['Inflação'].median()#Define a variável mediana contendo a mediana da inflação do dataframe criado.
media = marco['Inflação'].mean()#Define a variável media contendo a média da inflação do dataframe criado.
desvio = marco['Inflação'].std()#Define a variável desvio contendo o desvio-padrão da inflação do dataframe criado.
histo =  marco['Inflação'].hist()#Define a variável histo como o resultado do comando que cria o histograma dos dados de inflação do dataframe criado.
print (f'A mediana da inflação mensal nos meses de março após 1995 foi {mediana}\nA média da inflação mensal nos meses de março após 1995 foi {media}')
print (f'O desvio-padrão da inflação mensal nos meses de março após 1995 foi {desvio}\n\nHistograma dos meses de março após 1995:')
display (histo) #exibe os resultados, muito inferiores aos dados do exercício anterior pois o período selecionado exclui a época de hiperinflação
```

```python
#Gerador de números aleatórios
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
import numpy as np #importa a biblioteca numpy como np
N = [ 100 , 1000 , 10000, 100000]
rng = np.random.default_rng()
for item in N:
  x = rng.uniform(size=item)
  print (f'\nA média dos números gerados para N = {item} foi {np.mean(x)}\nHistograma:')
  plt.hist(x)
  plt.show()
```
