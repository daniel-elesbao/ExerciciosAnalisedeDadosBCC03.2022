```python
#Calcula a correlação entre renda e público do Campeonato Brasileiro 2018.
import pandas as pd #importa a biblioteca pandas como pd
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
brasileirao = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')#Lê o CSV fornecido pelo exercício
corr_publico_renda = brasileirao['Renda (R$)'].corr(brasileirao ['Público'])
r, p = np.polyfit(brasileirao['Renda (R$)'], brasileirao ['Público'], deg=1)
plt.figure( ) #executa função que desenha o gráfico
plt.title('Correlação entre Público e Renda no Campeonato Brasileiro de 2018') #Título do Gráfico
plt.plot (brasileirao['Renda (R$)'], brasileirao ['Público'], marker='o', color='red', linestyle='', label=f'Correlação: {corr_publico_renda:.2f}')#Cria o Gráfico da Renda pelo Público
plt.plot(brasileirao['Renda (R$)'], r*brasileirao['Renda (R$)'] + p)
plt.xlabel('Renda (milhões R$)') #define o eixo x como Renda
plt.ylabel('Público') #define o eixo y como Público
plt.legend ( )
plt.show( ) #exibe o gráfico

```
