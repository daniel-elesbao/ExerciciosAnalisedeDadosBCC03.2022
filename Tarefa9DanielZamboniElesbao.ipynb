```python
#Gráfico dos índices de analfabetismo em São Paulo, por ano, conforme dados disponibilizados por Ipeadata
import pandas as pd #importa a biblioteca pandas como pd
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa a biblioteca mathplotlib como plt
analfabetismo = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv')
plt.figure() #executa função que desenha o gráfico
plt.title('Índices de analfabetismo em São Paulo por ano')#Título do Gráfico
plt.plot (analfabetismo['Ano'],analfabetismo['São Paulo'])
plt.xlabel('Ano') #define o significado do eixo x como íons hidrônio
plt.ylabel('Índice de Analfabetismo em %') #define o significado do eixo y como Acidez
plt.show() #exibe o gráfico
```
