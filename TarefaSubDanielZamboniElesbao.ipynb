```python
#Modelo de simulação da interação entre infecções e vacinação em uma determinada população
import numpy as np #importa a biblioteca numpy como np
import matplotlib.pyplot as plt #importa mathplotlib como plt
def modelo_vacinacao ( vacdiaria, popinicial, dias ): #define a função modelo_vacinacao que recebe os valores vacinação diária, população inicial e dias
  alpha = 0.002 #dados de variação conforme fornecido pelo exercício
  beta = vacdiaria/popinicial #índice variável de acordo com o tamanho da população e o número diário de vacinados
  gamma = 0.15
  delta = 0.00005
  N = 3 
  P = np.zeros(dias) #cria as arrays para receber os dados e plotar os gráficos
  I = np.zeros(dias)
  C = np.zeros(dias)
  M = np.zeros(dias)
  datas = np.zeros(dias)
  P[0] = popinicial
  plt.figure( ) #executa função que desenha o gráfico
  for i in range (1, dias): #laço for que processa os cálculos pela quantidade de dias da simulação conforme definido pelo exercício
    P[i] = P[i-1] - (alpha * P[i-1]) - (beta * P[i-1]) + N
    I[i] = I[i-1] + (alpha * P[i-1]) - (gamma * I[i-1]) - (delta * I[i-1])
    C[i] = C[i-1] + (beta * P[i-1]) + (gamma * I[i-1])
    M[i] = M[i-1] + (delta * I[i-1])
    datas[i] = i-1
  plt.plot (datas, P, label=f'População Saudável não imunizada: {P[dias-1]:.0f}') 
  plt.plot (datas, I,  label=f'População Infectada: {I[dias-1]:.0f}') 
  plt.plot (datas, C, label=f'População imunizada: {C[dias-1]:.0f}') 
  plt.plot (datas, M, label=f'Mortes: {M[dias-1]:.0f}') #desenham as parábolas
  plt.plot (0, 14000) #estica o gráfico verticalmente com o fim de melhor posicionar as legendas
  plt.title ('Dados infectológicos de uma população')
  plt.xlabel ('Dias')
  plt.ylabel ('Pessoas')
  plt.grid ( ) #adiciona linhas para melhor visualização
  plt.legend ( loc = 1 ) #cria as legendas
  plt.show( ) #exibe o gráfico
modelo_vacinacao( 0, 10000, 365 ) # chama a função com os valores de vacinados por dia, população inicial e dias da simulação informados pelo exercício
#Para os dados informados o número de mortos não atinge 10 mesmo sem informação vacinados por dia
```
